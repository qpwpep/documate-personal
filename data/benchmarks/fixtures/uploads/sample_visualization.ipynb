# Visualization Notes
Histogram and pie chart examples with matplotlib.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.hist(values, bins=20, color='steelblue', edgecolor='white', alpha=0.8)
plt.title('Distribution')
plt.show()

labels = ['A', 'B', 'C']
sizes = [40, 35, 25]
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```
